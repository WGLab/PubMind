# Disease Extraction Benchmark
This is the main benchmark that will be shown in the PubMind paper for disease extraction performance. The benchmark dataset for disease NER is from NCBI disease corpus 34 used in PhenoTagger 35, which contains the PubMed abstract as well as the disease name found in the text.

```python
from bs4 import BeautifulSoup
import pandas as pd
from transformers import AutoTokenizer
import glob
from vllm import LLM, SamplingParams
import re
import sys

#filename_list_path='/mnt/isilon/wang_lab/shared/pubmed_central_decompressed/run1_1-90'


# Initialize the model
model = LLM('/mnt/isilon/wang_lab/pengwang/projects/LLM/Llama-3.3-70B-Instruct',
    tensor_parallel_size=2, # Change this to match the number of GPUs
    max_model_len=5000,
    gpu_memory_utilization=0.95,
    swap_space=0
)

```

```python
def generate_prompt(tmvar_corpus):
    
    prompts= []
    for paragraph in tmvar_corpus:
        prompt=f'{few_shot_examples}\nParagraph: \"{paragraph}\"'
        messages = [
            {'role': 'system', 'content': 
             '''You are an expert in genomics and bioinformatics. Extract the variant-related information from scientific publication paragraphs.'''},
            {'role': 'user', 'content': prompt},
        ]

        # Apply chat template
        templated_prompt = tokenizer.apply_chat_template(
            messages,
            add_generation_prompt=True,  # Adds necessary tokens for generation
            tokenize=False               # Returns a string
        )
        
        templated_prompt=templated_prompt.replace('Cutting Knowledge Date: December 2023\nToday Date: 26 Jul 2024\n\n','')
        prompts.append(templated_prompt)
    
    return prompts

# Function to convert model output to a DataFrame
def inference_to_df_old(output_test, pmc_ids):
    rows = []
    for i, x in enumerate(output_test):
        pmc_id = pmc_ids[i]  # Assign the correct PMC ID
        for line in x.outputs[0].text.strip().splitlines():
            if line.startswith("##"):
                parts = line[2:].split(", ")
                rows.append([pmc_id] + parts)
    
    df_cot = pd.DataFrame(rows)
    df_cot = df_cot.iloc[:,:7]
    #if df_cot.shape[1] == 7:  # Ensure the columns match the expected number
    df_cot.columns = ['PMCID', 'gene', 'DNA mutation', 'protein mutation', 'disease', 'experiment result', 'pathogenicity']
    df_cot = df_cot.drop_duplicates()
    df_cot = df_cot[(df_cot['DNA mutation'].str.contains(r'\d', na=False)) | 
                    (df_cot['protein mutation'].str.contains(r'\d', na=False))]
    return df_cot

# Function to convert model output to a DataFrame
def inference_to_df(output_test, pmc_ids):
    rows = []
    for i, x in enumerate(output_test):
        pmc_id = pmc_ids[i]  # Assign the correct PMC ID
        #sec = sec_for_inputs[i]
        for line in x.outputs[0].text.strip().splitlines():
            if "##" in line: # allow ## not at the beginning
                parts = line.split("##", 1)[-1].strip().split("||") 
                rows.append([pmc_id] + parts)
    
    df_cot = pd.DataFrame(rows)
    df_cot = df_cot.iloc[:,:8]
    #if df_cot.shape[1] == 7:  # Ensure the columns match the expected number
    df_cot.columns = ['PMID', 'gene', 'DNA mutation', 'protein mutation', 'disease', 'phenotype', 'LLM reasoning', 'pathogenicity']
    df_cot = df_cot.drop_duplicates()
    return df_cot
    
tokenizer = AutoTokenizer.from_pretrained("/mnt/isilon/wang_lab/pengwang/projects/LLM/Llama-3.3-70B-Instruct")
```

```python
# Load the corpus file
file_path = "/mnt/isilon/wang_lab/pengwang/projects/LLM/PhenoTagger/NCBItrainset_corpus.txt"

# Read the file line by line
with open(file_path, "r", encoding="utf-8") as file:
    lines = file.readlines()

# Initialize lists for storing data
abstract_data = []
annotation_data = []

# Process each line
for line in lines:
    line = line.strip()
    
    # Skip empty lines
    if not line:
        continue
    
    # Check if the line contains an abstract (|a| or |t|)
    if "|a|" in line or "|t|" in line:
        if "|a|" in line:
            parts = line.split("|")
            pmid = parts[0]  # PMID
            abstract = parts[-1]  # Abstract text
            abstract_data.append([pmid, abstract])
        #do not care about titile
        else:
            continue

    # Otherwise, it's an annotation line
    else:
        parts = line.split("\t")
        if len(parts) == 6:  # Ensure correct formatting
            pmid, start, end, entity, entity_type, disease_id = parts
            annotation_data.append([pmid, int(start), int(end), entity, entity_type, disease_id])

# Convert lists into DataFrames
df_abstracts = pd.DataFrame(abstract_data, columns=["PMID", "Abstract"])
df_annotations = pd.DataFrame(annotation_data, columns=["PMID", "Start", "End", "Entity", "Entity Type","Disease_ID"])
```

```python
all_inputs=list(df_abstracts['Abstract'])
final_inputs = generate_prompt(all_inputs)
pmc_ids_for_inputs=list(df_abstracts['PMID'])

print("Running inference ...")
# Run inference on all inputs
sampling_params = SamplingParams(temperature=0.6, top_p=0.3, max_tokens=1000)
output = model.generate(final_inputs, sampling_params)

```

```python
# Convert inference results to a single DataFrame
df_llmout = inference_to_df(output, pmc_ids_for_inputs)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Mean Pooling - Take attention mask into account for correct averaging
def meanpooling(output, mask):
    embeddings = output[0] # First element of model_output contains all token embeddings
    mask = mask.unsqueeze(-1).expand(embeddings.size()).float()
    return torch.sum(embeddings * mask, 1) / torch.clamp(mask.sum(1), min=1e-9)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Load model from HuggingFace Hub
pmbert_tokenizer = AutoTokenizer.from_pretrained("neuml/pubmedbert-base-embeddings")
pmbert_model = AutoModel.from_pretrained("neuml/pubmedbert-base-embeddings").to(device)

```

```python
device
```

```python
# Sentences we want sentence embeddings for
sentences = ['Hereditary hemochromatosis', 'hemochromatosis']

# Tokenize sentences
inputs = pmbert_tokenizer(sentences, padding=True, truncation=True, return_tensors='pt').to(device)

# Compute token embeddings
with torch.no_grad():
    output = pmbert_model(**inputs)

# Perform pooling. In this case, mean pooling.
embeddings = meanpooling(output, inputs['attention_mask'])

print("Sentence embeddings:")
print(embeddings)
```

```python
import torch.nn.functional as F
```

```python
# Compute cosine similarity
cos_sim = torch.cosine_similarity(embeddings[0].unsqueeze(0), embeddings[1].unsqueeze(0))

# Print the similarity score
print(f"Cosine Similarity: {cos_sim.item():.4f}")
```

```python
# Normalize the embeddings
embeddings_norm = F.normalize(embeddings, p=2, dim=1)  # L2 normalization

# Compute cosine similarity
cos_sim = torch.cosine_similarity(embeddings_norm[0].unsqueeze(0), embeddings_norm[1].unsqueeze(0))

# Print the similarity score
print(f"Cosine Similarity (norm): {cos_sim.item():.4f}")
```

```python
# Create a dictionary where the key is 'PMCID' and the value is a set of unique 'disease' values for each PMCID
pmcid_disease_dict = df_llmout.groupby('PMID')['disease'].apply(set).to_dict()

# Remove '-' from the sets in the dictionary
pmcid_disease_dict = {key: {disease for disease in value if disease.strip() != '-'} for key, value in pmcid_disease_dict.items()}

# Display the cleaned dictionary
pmcid_disease_dict
```

```python
# Create a dictionary where the key is 'PMCID' and the value is a set of unique 'disease' values for each PMCID
ncbi_disease_dict = df_annotations_dx.groupby('PMID')['Entity'].apply(set).to_dict()

# Remove '-' from the sets in the dictionary
ncbi_disease_dict = {key: {disease for disease in value if disease.strip() != '-'} for key, value in ncbi_disease_dict.items()}

# Display the cleaned dictionary
ncbi_disease_dict
```

```python
# Define a function to get embeddings using the provided mean pooling approach
def get_pubmedbert_embedding(text):
    inputs = pmbert_tokenizer(text, padding=True, truncation=True, return_tensors="pt").to(device)
    with torch.no_grad():
        output = pmbert_model(**inputs)
    return meanpooling(output, inputs["attention_mask"])

# Initialize a list to store results
similarity_results = []

# Iterate over each PMID in pmcid_disease_dict
for pmid, llm_diseases in pmcid_disease_dict.items():
    # Get corresponding diseases from ncbi_disease_dict
    ncbi_diseases = ncbi_disease_dict.get(pmid, set())

    # Skip if no corresponding diseases in NCBI dictionary
    if not llm_diseases or not ncbi_diseases:
        continue  # Skip if either is empty

    # Compute embeddings for LLM and NCBI diseases using PubMedBERT
    llm_embeddings = torch.cat([get_pubmedbert_embedding(disease) for disease in llm_diseases])
    ncbi_embeddings = torch.cat([get_pubmedbert_embedding(disease) for disease in ncbi_diseases])

    # Compute cosine similarity between each LLM disease and all NCBI diseases
    cosine_similarities = torch.nn.functional.cosine_similarity(
        llm_embeddings.unsqueeze(1), ncbi_embeddings.unsqueeze(0), dim=2
    )

    # Store the max similarity for each LLM disease
    for i, llm_disease in enumerate(llm_diseases):
        max_similarity, max_index = torch.max(cosine_similarities[i], dim=0)
        most_similar_ncbi_disease = list(ncbi_diseases)[max_index.item()]
        similarity_results.append([pmid, llm_disease, max_similarity.item(), most_similar_ncbi_disease])

# Convert results to a DataFrame
df_similarity = pd.DataFrame(similarity_results, columns=["PMID", "Disease_from_LLM", "Max_Similarity_Score", "Most_Similar_NCBI_Disease"])
df_similarity
```
